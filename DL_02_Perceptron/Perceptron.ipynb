# Rough Model

```python
from sklearn.metrics import accuracy_score
```

```python
w=np.ones(X_train.shape[1]) #intial weights =1
b=0 #initial threshold
```

```python
for i in range(100): #Iterating 100 times over data
    for x,y in zip(X_train,Y_train):
        if np.dot(x,w) < b:
            y_pred= 0
        else:
            y_pred=1
        if y == 1 and y_pred == 0:
            w=w+ x*.005 #setting learning rate .005
            b=b- 1*.005
        if y == 0 and y_pred == 1:
            w=w- x*.005
            b=b+ 1*.005
```

```python
y_pred_test=[]
for x in X_test:
    if np.dot(x,w) < b:
        y_pred= 0
    else:
        y_pred=1
    y_pred_test.append(y_pred)
```

```python
print(y_pred_test)
```

```python
accuracy= accuracy_score(y_pred_test,Y_test)
print(accuracy)
```

# Perceptron Neuron

```python
from sklearn.metrics import accuracy_score
```

```python
class Perceptron:

    def __init__(self):
        self.b=None
        self.w=None
    
    def  model(self, x):
        return 1 if np.dot(x,self.w) >= self.b else 0

    def predict(self, X):
        y=[]
        for x in X:
            pred= self.model(x)
            y.append(pred)
        return np.array(y)

    def fit(self, X, Y, epochs=1, lr=1):
        best_accuracy= 0
        self.b= 0
        self.w= np.zeros(X.shape[1])
        best_b= None
        best_w= None
        for i in range(epochs):
            for x,y in zip(X,Y):
                pred= self.model(x)
                if y==1 and pred == 0:
                    self.w= self.w + x*lr
                    self.b= self.b - 1*lr
                elif y==0 and pred == 1:
                    self.w= self.w - x*lr
                    self.b= self.b + 1*lr
       
            accuracy= accuracy_score(self.predict(X),Y)
            if accuracy > best_accuracy:
                best_accuracy= accuracy
                best_b= self.b
                best_w= self.w

        self.b=best_b
        self.w= best_w 
```

```python
perceptron= Perceptron()
perceptron.fit(X_train,Y_train,1000,.005)
```

```python
pred=perceptron.predict(X_test)
```

```python
accuracy= accuracy_score(pred,Y_test)
print(accuracy)
```
